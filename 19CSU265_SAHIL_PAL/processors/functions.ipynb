```python
# import packages
from IPython.display import clear_output
import time
import os
import emoji
import interface

# method to implement the functionality of adding task
def add_task(task_name , task_type , task_time, username):
    task_time = str(task_time)
    temp = "Task Name : "+task_name+"   Task Type : "+task_type+"   Task Time :  "+task_time+"  hrs"
    extention = ".txt"
    filedata = username+extention
    file = open(filedata, 'a') 
    file.write("\n"+temp)
    file.close()
    interface.add_task()
    
def display_task(username):
    interface.display_task()
    extention = ".txt"
    filedata = username+extention
    file = open(filedata, 'r') 
    for each in file:
        if each == "\n":
            print("            |                                                                                       |")
        else:
            li = list(each.split("   "))
            print("            |",format(li[0], '>2'),format(li[1], '>25'),format(li[2], '>20'))
            print("            |                                                                                       |")
    file.close()
    print("             --------------------------------------------------------------------------------------- ")
    
    # choice for going to main menu
    choice = input("              < Do you want to continue ... [Y : yes] >")
    if(choice == 'Y' or choice == 'y'):
        interface.loading()
        return 'y'
    else:
        print("             <  Thanks for using... --> ( Exiting in 2 secs...) > ")
        time.sleep(2)
        interface.loading()
        os._exit(1)

def update_task(task_name, updated_taskname, updated_tasktype, updated_tasktime, username):
    task_name = "Task Name : "+task_name
    update_task = "Task Name : "+updated_taskname+"   Task Type : "+updated_tasktype+"   Task Time :  "+str(updated_tasktime)+"  hrs"
    extention = ".txt"
    filedata = username+".txt"
    filetemp = "temp"+username+".txt"
    file1 = open(filedata, 'r')
    file2 = open(filetemp, 'w')
    for each in file1:
        li = list(each.split("   "))
        if li[0] == task_name:
            file2.write(update_task+"\n")
        else:
            file2.write(each)
    file2.close()
    file1.close()
    os.remove(filedata)
    os.rename(filetemp,filedata)
    interface.update_task()
    

def delete_task(task_name,username):
    task_name = "Task Name : "+task_name
    extention = ".txt"
    filedata = username+".txt"
    filetemp = "temp"+username+".txt"
    file1 = open(filedata, 'r')
    file2 = open(filetemp, 'w')
    for each in file1:
        li = list(each.split("   "))
        if li[0] == task_name:
            print()
        else:
            file2.write(each)
    file2.close()
    file1.close()
    os.remove(filedata)
    os.rename(filetemp,filedata)
    interface.delete_task()
    
def calc_productivity(username):
    total_hours = 0.0
    prod_hours = 0.0 
    extention = ".txt"
    filedata = username+extention
    file = open(filedata, 'r') 
    for each in file:
        if each == "\n":
            print()
        else:
            li = list(each.split("   "))
            li_new = list(li[2].split("  "))
            if(li[1] == "Task Type : productive"):
                prod_hours = float(li_new[1]) + prod_hours
                total_hours = float(li_new[1]) + total_hours
            else:
                total_hours = float(li_new[1]) + total_hours
    file.close()
    
    productivity_percent = (prod_hours/total_hours)*100
    clear_output()
    print("             ---------------------------------------------------------------------------------------")
    print("            |                             _________________________                                 |")
    print("            |                            | PRODUCTIVITY CALCULATOR |                                |")
    print("            |                             -------------------------                                 |")
    print("            |                             CALCULATING PRODUCTIVITY...                               |")
    print("             --------------------------------------------------------------------------------------- ")
    print("            |                                                                                       |")
    if(productivity_percent <= 50):
        print("            |",username," your productivity is : ",format(productivity_percent,'.2f')," \U0001F971                                               |")
        print("            |             ---> Your productive hours : ",str(prod_hours),"                                         |")
        print("            |             ---> Your total hours of task : ",str(total_hours),"                                     |")
        print("            |                                                                                       |")
        print("            |                                                                                       |")
        print("             ---------------------------------------------------------------------------------------")
    else:
        print("            |",username,",  your productivity is : ",format(productivity_percent,'.2f')," \U0001F603                                         |")
        print("            |             ---> Your productive hours : ",str(prod_hours),"                                        |")  
        print("            |             ---> Your total hours of task : ",str(total_hours),"                                    |")
        print("            |                                                                                       |")
        print("            |                                                                                       |")
        print("             ---------------------------------------------------------------------------------------")
    print("")
    choice = input("              < Do you want to continue ... [Y : yes] >")
    if(choice == 'Y' or choice == 'y'):
        return 'y'
    else:
        print("             <  Thanks for using... --> ( Exiting in 2 secs...) > ")
        time.sleep(2)
        interface.loading()
        os._exit(1)
```
