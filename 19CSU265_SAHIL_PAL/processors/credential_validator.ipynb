```python
# importing packages
from IPython.display import clear_output
import re

# method to validate username for new user
def username_check(username):
    pattern = '[a-zA-z0-9]'
    length = len(username)
    x = re.search(pattern,username)
    if x and length >= 4 and length < 15:
        return True
    else:
        return False


# method to validate password for new user
def password_check(password):
    pattern = '^[a-zA-z0-9/D]'
    length = len(password)
    x = re.findall(pattern, password)
    if x and length >= 8 and length < 20:
        return True
    else:
        return False
    

# re-directing invalid username page
def redirect_username():
    clear_output()
    print("             ---------------------------------------------------------------------------------------")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                             _________________________                                 |")
    print("            |                            | PRODUCTIVITY CALCULATOR |                                |")
    print("            |                             -------------------------                                 |")
    print("            |                                  New User Portal                                      |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("             --------------------------------------------------------------------------------------- ")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                              < NEW USER CREATION FAILED >                             |")
    print("            |                                                                                       |")
    print("            |                             < ENTERED USERNAME IS INVALID >                           |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                              Redirecting to the sign up page....      |")
    print("             --------------------------------------------------------------------------------------- ")

# re-directing invalid password page
def redirect_password():
    clear_output()
    print("             ---------------------------------------------------------------------------------------")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                             _________________________                                 |")
    print("            |                            | PRODUCTIVITY CALCULATOR |                                |")
    print("            |                             -------------------------                                 |")
    print("            |                                  New User Portal                                      |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("             --------------------------------------------------------------------------------------- ")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                              < NEW USER CREATION FAILED >                             |")
    print("            |                                                                                       |")
    print("            |                             < ENTERED PASSWORD IS INVALID >                           |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                              Redirecting to the sign up page....      |")
    print("             --------------------------------------------------------------------------------------- ")
```
