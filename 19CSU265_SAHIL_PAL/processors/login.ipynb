```python
# importing packages
from IPython.display import clear_output
import time


# checking user credential from credential.txt and returning as per check
def login_validator(username , password):
    temp = username + password
    result = False
    file = open('credentials.txt', 'r') 
    for each in file:
        if(each == temp):
            result = True
        else:
            print()
    return result


#re-directing page for failed login
def redirect_display():
    clear_output()
    print("             ---------------------------------------------------------------------------------------")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                             _________________________                                 |")
    print("            |                            | PRODUCTIVITY CALCULATOR |                                |")
    print("            |                             -------------------------                                 |")
    print("            |                                    LOGIN PAGE                                         |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("             --------------------------------------------------------------------------------------- ")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                  < LOGIN FAILED >                                     |")
    print("            |                                                                                       |")
    print("            |                            < CHECK USERNAME AND PASSWORD >                            |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                   Redirecting to the home page....    |")
    print("             --------------------------------------------------------------------------------------- ");
    


# login display page 
def display():
    clear_output()
    print("             ---------------------------------------------------------------------------------------")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                             _________________________                                 |")
    print("            |                            | PRODUCTIVITY CALCULATOR |                                |")
    print("            |                             -------------------------                                 |")
    print("            |                                    LOGIN PAGE                                         |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("             --------------------------------------------------------------------------------------- ")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    username = input("            |    Enter Username   ----   ")                       
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    password = input("            |    Enter Password   ----   ")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("             --------------------------------------------------------------------------------------- "); 
   
    # calling the credential checker
    check = login_validator(username , password)
    
    # return appropriately to main program for the program to run furthur
    if(check == True):
        return 'True',username
    else:
        redirect_display()
        time.sleep(3)
        return 'False',username
        
    

```
