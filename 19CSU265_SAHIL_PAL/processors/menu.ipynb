```python
# importing packages
from IPython.display import clear_output
import import_ipynb
import functions,interface,my_exceptions
import time,os

# few predefined task using dictionary
predefined_type = {'sleep':'essential' ,'eat' : 'essential','watching TV': 'non-productive',
                   'cycling':'productive','gyming' : 'productive', 'reading books': 'productive', 
                   'reading newspaper': 'productive','travel': 'non-producive', 'talking':'non-productive' ,
                   'sports':'productive', 'teaching':'productive','studying':'productive', 
                   'shopping': 'non-productive', 'cleaning':'productive', 'research':'productive', 
                   'bath' : 'essential' , 'medicine' : 'essential'}

temp_username = ""

# menu display with choices
def display(username):
    temp_username = ""+username
    terminate_display = False
    while terminate_display == False:
        clear_output()
        print("             --------------------------------------------------------------------------------------- ");
        print("            |                            | PRODUCTIVITY CALCULATOR |                                |");
        print("            |                             -------------------------                                 |");
        print("            |                                    Dash Board                                         |");
        print("             ---------------------------------------------------------------------------------------");
        print("            |                   ________________________                                            |");
        print("            |    --- Press[1]  | ADD TASK               |                                           |");
        print("            |                   ------------------------                                            |");
        print("            |                   ________________________                                            |");
        print("            |    --- Press[2]  | CALCULATE PRODUCTIVITY |                                           |");
        print("            |                   ------------------------                                            |");
        print("            |                   ________________________                                            |");
        print("            |    --- Press[3]  | UPDATE TASK            |                                           |");
        print("            |                   ------------------------                                            |");
        print("            |                   ________________________                                            |");
        print("            |    --- Press[4]  | DELETE TASK            |                                           |");
        print("            |                   ------------------------                                            |");
        print("            |                   ________________________                                            |");
        print("            |    --- Press[5]  | DISPLAY TASK           |                                           |");
        print("            |                   ------------------------                                            |");
        print("            |                   ________________________                                            |");
        print("            |    --- Press[6]  | EXIT                   |                                           |");
        print("            |                   ------------------------                                            |");
        print("             ---------------------------------------------------------------------------------------");
        function_choice = int(input("                 [ Enter choice here ]   ----   "))
        
        if(function_choice == 1):
            add(temp_username)
            clear_output()
        elif(function_choice == 3):
            update(temp_username)
            clear_output()
        elif(function_choice == 5):
            disp(temp_username)
        elif(function_choice == 2):
            calc_pro(temp_username)
            clear_output()
        elif(function_choice == 4):
            delete(temp_username)
            clear_output()
        elif(function_choice == 6):
            print("                 <  THANKS FOR USING  > ")
            print("             < Logging out and exiting in 2 secs >")
            time.sleep(2)
            interface.loading()
            os._exit(1)
            

# input for adding a task
def add(temp_username):
    clear_output()
    task_name =""
    task_type =""
    task_time = 0.0
    print("             --------------------------------------------------------------------------------------- ");
    print("            |                                                                                       |");
    print("            |                             _________________________                                 |");
    print("            |                            | PRODUCTIVITY CALCULATOR |                                |");
    print("            |                             -------------------------                                 |");
    print("            |                                     Add Task                                          |");
    print("             ---------------------------------------------------------------------------------------");
    print("            |                                                                                       |")
    task_name = input("            |    Enter Task Name   ----   ")  
    print("            |                                                                                       |");
    
    terminate_tasktype = False
    task_type=""
    if(task_name in predefined_type):
        task_type = predefined_type[task_name]
    else:
        print("            |          < Productive / non-productive / essential >                                              |")
        while terminate_tasktype == False:
            task_type = input("            |    Enter Task Type   ----   ")
            option = {'productive','non-productive','essential'}
            # throwing exception for invalid task_type
            try:
                if task_type not in option:
                    raise my_exceptions.typeError("  TYPE_ERROR")
                else :
                    terminate_tasktype = True
            except my_exceptions.typeError as e:
                print(e,"Please re-enter...")
           
    print("            |                                                                                       |");    
    time_terminate = False
    while time_terminate == False:
        task_time = float(input("            |    Enter Time Spent   ----   ")) 
        # throwing exception for invalid task_time
        try:
            if not 0 < task_time <= 24:
                raise my_exceptions.timeError(time)
            else:
                time_terminate = True
        except my_exceptions.timeError as e:
            time_terminate = False
            print(e,"Please re- enter....")
    
    
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("             --------------------------------------------------------------------------------------- ")
    
    # calling the add functionality
    user = temp_username
    functions.add_task(task_name, task_type , task_time, user)
    clear_output()
    display(temp_username)
    
# input for updating the task    
def update(temp_username):
    clear_output()
    task_name =""
    updated_taskname = ""
    updated_tasktype = ""
    updated_tasktime = ""
    print("             --------------------------------------------------------------------------------------- ");
    print("            |                                                                                       |");
    print("            |                             _________________________                                 |");
    print("            |                            | PRODUCTIVITY CALCULATOR |                                |");
    print("            |                             -------------------------                                 |");
    print("            |                                     Update Task                                       |");
    print("             ---------------------------------------------------------------------------------------");
    print("            |                                                                                       |")
    task_name = input("            |    Enter Task Name to be updated  ----   ")  
    print("            |                                                                                       |");
    print("            |           < Enter 'N / n / 0' in the field you don't want to update >                 |")
    updated_taskname = input("            |    Enter Task Name   ----   ")  
    print("            |                                                                                       |");
    
    terminate_tasktype = False
    if(updated_taskname in predefined_type):
        updated_tasktype = predefined_type[updated_taskname]
    else:
        print("            |          < Productive / non-productive / essential >                                              |")
        while terminate_tasktype == False:
            updated_tasktype = input("            |    Enter Task Type   ----   ")
            option = {'productive','non-productive','essential'}
            # throwing exception for invalid task_type
            try:
                if updated_tasktype not in option:
                    raise my_exceptions.typeError("  TYPE_ERROR")
                else :
                    terminate_tasktype = True
            except my_exceptions.typeError as e:
                print(e,"Please re-enter...")
  
    print("            |                                                                                       |");    
    
    time_terminate = False
    while time_terminate == False:
        updated_tasktime = float(input("            |    Enter Time Spent   ----   ")) 
        # throwing exception for invalid task_time
        try:
            if not 0 < updated_tasktime <= 24:
                raise my_exceptions.timeError(time)
            else:
                time_terminate = True
        except my_exceptions.timeError as e:
                time_terminate = False
                print(e,"Please re- enter....")
    
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("             --------------------------------------------------------------------------------------- ")
    
    # calling the update functionalty
    user = temp_username
    functions.update_task(task_name, updated_taskname, updated_tasktype , updated_tasktime , user)
    clear_output()
    display(temp_username)
    

# display function call
def disp(temp_username):
    user = temp_username
    check = functions.display_task(user)
    if(check == 'y'):
        clear_output()
        display(temp_username)
    else:
        clear_output()

# calculate productivity function call
def calc_pro(temp_username):
    user = temp_username
    check = functions.calc_productivity(user)
    if(check == 'y'):
        clear_output()
        display(temp_username)
    else:
        clear_output()


# input for deleting a task
def delete(temp_username):
    clear_output()
    task_name =""
    print("             --------------------------------------------------------------------------------------- ");
    print("            |                                                                                       |");
    print("            |                             _________________________                                 |");
    print("            |                            | PRODUCTIVITY CALCULATOR |                                |");
    print("            |                             -------------------------                                 |");
    print("            |                                    Delete Task                                        |");
    print("             ---------------------------------------------------------------------------------------");
    print("            |                                                                                       |")
    task_name = input("            |    Enter Task Name to be deleted  ----   ")  
    print("            |                                                                                       |");
    # call to input functionality
    user = temp_username
    functions.delete_task(task_name,user)
    clear_output()
    display(temp_username)
    
```
