```python
# importing packages
from IPython.display import clear_output
import import_ipynb
import credential_validator
import time

# sign up page 
def display():
    clear_output()
    print("             ---------------------------------------------------------------------------------------")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                             _________________________                                 |")
    print("            |                            | PRODUCTIVITY CALCULATOR |                                |")
    print("            |                             -------------------------                                 |")
    print("            |                                 New User Portal                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("             --------------------------------------------------------------------------------------- ")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |   < Username must be in between 4 - 15 and must start with non - special character >  |")
    username = input("            |    Enter Username   ----     ")  
    print("            |                                                                                       |")
    print("            |   < Password must be in between 8 - 20 >                                              |")
    print("            |   < must contain one uppercase character and digit >                                  |")
    print("            |                                                                                       |")
    password = input("            |    Enter Password   ----     ")    
    print("            |                                                                                       |")
    cpassword = input("            |    Confirm password   ----    ")  
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("             --------------------------------------------------------------------------------------- ") 
    
    if(password == cpassword):
        # calling validations of username and password
        if(credential_validator.username_check(username) == True):
            if(credential_validator.password_check(username) == True):
                user_creater(username , password)
                return 'True',username
            else:
                credential_validator.redirect_username()
                time.sleep(4)
                clear_output()
                display()
                return 'False',username
        else:
            credential_validator.redirect_password()
            time.sleep(4)
            clear_output()
            display()
            return 'False',username
    else:
        redirect_display()
        time.sleep(4)
        clear_output()
        display()
        return 'False',username
        

# adding details to credential file and creating a seprate data file for each user
def user_creater(username , password):
    temp = username + password
    filename = username+".txt"
    file = open('credentials.txt', 'a') 
    filenew = open(filename,'w')
    file.write("\n")
    file.write(temp)
    filenew.close()
    file.close()

# re-directing display page 
def redirect_display():
    clear_output()
    print("             ---------------------------------------------------------------------------------------")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                             _________________________                                 |")
    print("            |                            | PRODUCTIVITY CALCULATOR |                                |")
    print("            |                             -------------------------                                 |")
    print("            |                                  New User Portal                                      |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("             --------------------------------------------------------------------------------------- ")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                              < NEW USER CREATION FAILED >                             |")
    print("            |                                                                                       |")
    print("            |               < ENTERED PASSWORD AND CONFIRMED PASSWORD DID NOT MATCHED >             |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                                                                       |")
    print("            |                                              Redirecting to the sign up page....      |")
    print("             --------------------------------------------------------------------------------------- ")
    
    
    
```
